```python
import seaborn as sns
import matplotlib as plt
import pandas as pd
```

```python
# I WANNA DATA CLEAN BUT HUHU CLEAN NA DAAN PIEN so visual nalang
# Training data
train_df =  pd.read_csv("../data/csv/train_data.csv")
dogs = train_df[train_df['animal'] == 0]
print("Dogs size: ",dogs.size)
cats = train_df[train_df['animal'] == 1]
print("Cats size: ",cats.size)
sns.countplot(data = train_df, x = "animal").set(title = "Training Data")
```

```python
test_df =  pd.read_csv("../data/csv/test_data.csv")
dogs = test_df[test_df['animal'] == 0]
print("Dogs size: ",dogs.size)
cats = test_df[test_df['animal'] == 1]
print("Cats size: ",cats.size)
sns.countplot(data = test_df, x = "animal").set(title = "Test Data")
```
